```python
#Arithmetic operations
a1 = np.array([1,2,3,4,5,6])
a2 = np.array([7,8,9,10,11,12])
a1 + a2
a1 * a2
```

```python
#Broadcasting
array = np.array([1,2,3,4,5])
array + 20
```

```python
#matrix mul
a1 = np.array([[1,2],[3,4]])
a2 = np.array([[5,6],[7,8]])

a1@ a2
np.dot(a1,a2)
```

```python
#transpose 
a1.T
```

```python
#stacking an array
a3 = np.array([1,2,3])
a4 = np.array([4,5,6])
np.vstack((a3,a4))
np.hstack((a3,a4))
np.column_stack((a3,a4))
```
