```python
df = pd.read_csv('IPL.csv')
df.head()
```

```python
df.shape
```

```python
match_wins = df['match_winner'].value_counts()
sns.barplot(y = match_wins.index,x=match_wins.values,palette='viridis')
plt.title("Match wins by each team")
```

```python
sns.countplot(x = df['toss_decision'],palette="rainbow")
plt.title("Toss decision")
```

```python
count = df[df["toss_winner"] == df["match_winner"]]['match_id'].count()
percentage = (count*100)/df.shape[0]
percentage.round(2)
```

```python
count = df['player_of_the_match'].value_counts().head(5)
count
```

```python
sns.barplot(x=count.values,y = count.index,palette='magma')
plt.title("Most player of the match")
```

```python
df.head()
```

```python

high = df.groupby('top_scorer')['highscore'].sum().sort_values(ascending=False).head(2)
high
```

```python
high.plot(kind='bar',color=['blue','orange'])
plt.title("Top score in IPL")
```

```python
best_bow = df.groupby('best_bowling')['best_bowling_figure'].sum().sort_values(ascending=False).head(2)
best_bow
```

```python
df["highest_wicket"] = df["best_bowling_figure"].apply(lambda x : x.split('--')[0])
df
```

```python
df["highest_wicket"] = df["highest_wicket"].astype(int)
```

```python
best_bow = df.groupby('best_bowling')['highest_wicket'].sum().sort_values(ascending=False).head(10)
best_bow
best_bow.plot(kind='bar',color ='hotpink')
plt.title("Best Boweler")
```

```python
venue = df['venue'].value_counts()
venue
venue.plot(kind='bar',color="yellow")
plt.title("Most played match stadium")
```
