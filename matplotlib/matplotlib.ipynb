```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from random import sample
import matplotlib.image as mpimg
```

```python
x = np.linspace(0,5,11)
x
y = x ** 2
y
```

```python
plt.plot(x,y)
```

```python
plt.subplot(2,2,1)
plt.plot(y,x)
plt.subplot(2,2,2)
plt.plot(y**2,y**2)
plt.subplot(2,2,3)
plt.plot(x**2,x**2)
plt.subplot(2,2,4)
plt.plot(x,y)
```

```python
fig = plt.figure()
axis1 = fig.add_axes([0.1,0.1,1,1])
axis1.plot(x,y)
axis2 = fig.add_axes([0.3,0.4,0.3,0.3])
axis2.plot(x,y)
```

```python
plt.scatter(x,y)
```
