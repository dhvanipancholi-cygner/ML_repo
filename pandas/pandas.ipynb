```python
# merging two dataframes

employees = pd.DataFrame({
    'emp_id': [1, 2, 3, 4,7],
    'name': ['Alice', 'Bob', 'Charlie', 'David','eve'],
    'department': ['HR', 'IT', 'Finance', 'Marketing','Sales']
})

departments = pd.DataFrame({
    'emp_id': [1, 2, 3, 4,5,6],
    'dept_name': ['Human Resources', 'Information Technology', 'Finance', 'Marketing','Sales','BDE']
})
```

```python
employees
```

```python
departments
```

```python
pd.merge(employees,departments)
```

```python
pd.merge(employees,departments,on="emp_id",how= 'inner')
```

```python
pd.merge(employees,departments,on="emp_id",how= 'outer')
```

```python
pd.merge(employees,departments,on="emp_id",how= 'left')
```

```python
pd.merge(employees,departments,on="emp_id",how= 'right')
```
